````python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(42)
data = pd.DataFrame({
    'TransactionAmount': np.random.uniform(10, 1000, 100),
    'AccountBalance': np.random.uniform(500, 5000, 100),
    'TransactionFrequency': np.random.poisson(5, 100)
})

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Choose the number of clusters (you may want to experiment with this)
num_clusters = 3

# Apply KMeans clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
data['Cluster'] = kmeans.fit_predict(scaled_data)

# Visualize the clusters (for two features)
plt.scatter(data['TransactionAmount'], data['AccountBalance'], c=data['Cluster'], cmap='viridis')
plt.title('Customer Segmentation')
plt.xlabel('Transaction Amount')
plt.ylabel('Account Balance')
plt.show()

# Display the cluster centers (in the standardized feature space)
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
cluster_centers_df = pd.DataFrame(cluster_centers, columns=data.columns[:-1])
print(cluster_centers_df)
````

### Introduction to Cluster Evaluation Techniques

In the realm of unsupervised machine learning, determining the optimal number of clusters is a pivotal decision that can significantly impact the outcomes of your model. Cluster evaluation techniques are essential tools that provide guidance in this decision-making process. Two of the most widely recognized methods for evaluating clustering results are the Elbow Method and the Silhouette Score.

#### Elbow Method
- **Explanation**: The Elbow Method is a heuristic used in determining the number of clusters in a data set. The approach involves plotting the explained variance as a function of the number of clusters, and picking the point where the increase in variance explained by adding another cluster is not significant anymore. This point is known as the 'elbow', where the graph bends.
- **Interpretation**: In the Elbow Method, one should look for a change in the gradient of the line plot; a sharp change like an elbow suggests the optimal number of clusters. The idea is that adding more clusters beyond this number does not provide much better modeling of the data.

#### Silhouette Score
- **Explanation**: The Silhouette Score is a metric used to calculate the goodness of a clustering technique. It measures the distance between points within a cluster and the distance to points in the next nearest cluster. The score ranges from -1 to +1, where a high value indicates that the points are well clustered.
- **Interpretation**: A Silhouette Score close to +1 indicates that the sample is far away from the neighboring clusters. A value of 0 indicates that the sample is on or very close to the decision boundary between two neighboring clusters and negative values indicate that those samples might have been assigned to the wrong cluster.

Both methods provide different lenses through which to view the clustering results and can be used in conjunction to make a more informed decision. The Elbow Method gives us an insight into the variance within each cluster, whereas the Silhouette Score provides a measure of how similar an object is to its own cluster compared to others. The optimal number of clusters is often the one that balances between the two measures, subject to the specific context and use case of the analysis.

````python
# Step 2: Enhanced clustering with K-Means
# standardizing the data (excluding non-numeric columns)
features = data[['Annual Income (k$)', 'Spending Score (1-100)']]
scaler = StandardScaler()
scaled_data = scaler.fit_transform(features)
````

````python
#Calculate KMeans with different cluster numbers
# Elbow Method
inertia = []
for i in range(1, 11):
  kmeans = KMeans(n_clusters=i, random_state = 42)
  kmeans.fit(scaled_data)
  inertia.append(kmeans.inertia_)

# Plotting the Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method to Determine Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
````

````python
#Calculate Kmeans with different Cluster numbers
#Silhouette Score
silhouette_scores = []
for i in range(2,11):
  kmeans = KMeans(n_clusters=i, random_state=42)
  kmeans.fit(scaled_data)
  labels = kmeans.labels_
  score = silhouette_score(scaled_data,labels)
  silhouette_scores.append(score)

#Plotting silhouette score
plt.figure(figsize=(8,5))
plt.plot(range(2,11), silhouette_scores, marker='o')
plt.title('Silhouette Score to Determine Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()
````

**Observations from K-Means Clustering**
1. **Elbow Method**
  * The **Elbow Method** suggests an optimal number of clusters around 5, where the inertia begins to level off.

2. **Silhouette Score**
  * The **Silhouette Score** peaks at 5, this suggests the taking 5 clusters into consideration is a suitable option, as it provides more granular segmentation.
  * To justify, The silhouette score for 5 clusters is quite high compared to the others, indicating well-seperated clusters. This number of clusters also aligns with potential customer segments based on annual income and spending behaviour.

* Comparing both Elbow method and Silhouette score choosing **5 clusters** balances between model complexity and interpretability, capturing distinct customer segments without overfitting the model.

## Advance Clusterting Techniques
**DBSCAN Clustering Implementation**
  * DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a clustering method that groups together closely packed points (those with many nearby neighbors) and marks points that lie alone in low-density regions as outliers.

Let's implement DBSCAN and compare its clustering results with K-Means and Hierarchical Clustering.

````python
# Importing necessary libraries
from sklearn.cluster import DBSCAN

# Initializing DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(scaled_data)

# Adding DBSCAN labels to the original data
data['DBSCAN_Cluster'] = dbscan_labels
````

````python
# Calculate the number of clusters and outliers
unique_labels = np.unique(dbscan_labels)
n_clusters_dbscan = len(unique_labels) - (1 if -1 in dbscan_labels else 0)
n_outliers = list(dbscan_labels).count(-1)

print(f"Number of Clusters (DBSCAN): {n_clusters_dbscan}")
print(f"Number of Outliers (DBSCAN): {n_outliers}")
````

````python
# Visualization for DBSCAN Clustering
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['Annual Income (k$)'], y=data['Spending Score (1-100)'], hue=data['DBSCAN_Cluster'], palette='viridis')
plt.title('DBSCAN Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='DBSCAN Cluster', loc='upper right')
plt.show()
````

## **DBSCAN Clustering Insights:**

Number of Clusters: 2 clusters

Number of Outliers: 8 outliers identified (represented by -1 in the plot)

**Cluster Characteristics:**
  * **Cluster 0:** This cluster contains the majority of customers, spread across a wide range of income levels and spending scores. It is more densely populated, particularly around moderate income and spending levels.
  * **Cluster 1:** This cluster captures high-spending customers with relatively higher income. It’s smaller but more concentrated.
  * **Outliers:** DBSCAN identifies several outliers, scattered across the income and spending spectrum. These outliers represent customers who do not fit neatly into any cluster, potentially indicating unique or anomalous spending behavior.

## **Dimensionality Reduction Technique**
**Principal Component Analysis Implementation:**

PCA is a dimensionality reduction technique that transforms a dataset into a lower-dimensional space while preserving as much variance as possible. This can be useful for visualizing high-dimensional data and potentially improving clustering performance.

````python
from sklearn.decomposition import PCA

# Apply PCA to reduce dimensions
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

# Create a DataFrame with the PCA results
data['PCA1'] = pca_result[:, 0]
data['PCA2'] = pca_result[:, 1]

# Visualize the PCA results
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['PCA1'], y=data['PCA2'], hue=data['Cluster'], palette='viridis')
plt.title('PCA Results Colored by K-Means Clusters')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.show()
````


## **Impact of Dimensionality Reduction on Clustering**

1. **Cluster Separation:**

    * **PCA Visualization:** The plot shows that the clusters identified by K-Means are fairly well-separated in the reduced two-dimensional space. This suggests that the principal components have successfully captured the key variations in the data, allowing us to visualize the clustering in a simplified form.
    * **Interpretation:** PCA has maintained the integrity of the clusters, which indicates that the original features contributing to these clusters are well-represented by the first two principal components.
2. **Noise Reduction:**

    * **Dimensionality Reduction:** By reducing the dataset to just two dimensions, PCA helps eliminate noise or less informative features that might clutter the clustering process. This can lead to clearer and more distinct clusters, as observed in the plot.
    * **Impact on Clustering:** The more noise or redundant features in the original dataset, the more beneficial PCA becomes. In this case, it seems to have clarified the clusters without losing important information.

3. **Simplification of Data:**

    * **Ease of Visualization:** Reducing the data to two dimensions makes it much easier to visualize the relationships between clusters. This is particularly useful when dealing with high-dimensional data, where direct visualization is not possible.
    * **Data Interpretation:** Simplifying the dataset using PCA allows for more straightforward interpretation of the clustering results, making it easier to explain the segments to stakeholders who may not have technical expertise.


## **Potential Use of PCA in Simplifying Complex Datasets**
1. **Dimensionality Reduction in High-Dimensional Spaces:**

    * **Real-World Scenarios:** In datasets with many features (e.g., customer data with numerous behavioral and demographic attributes), PCA can reduce the number of dimensions while retaining the most significant information.
    * **Efficiency:** Reducing dimensions can significantly improve computational efficiency, especially in large datasets, by lowering the amount of data that needs to be processed.
2. **Improving Clustering Results:**

    * **Feature Importance:** PCA helps focus the clustering process on the most important features, which can lead to more meaningful clusters. This is particularly valuable when some features might introduce noise or irrelevant variations.
    * **Pre-Processing Step:** PCA can be used as a pre-processing step before applying clustering algorithms, particularly when the original dataset contains highly correlated or redundant features.
3. **Enhanced Data Exploration:**

    * **Exploratory Data Analysis:** PCA is a powerful tool for exploratory data analysis (EDA), allowing analysts to identify patterns, trends, and outliers in the data that might not be immediately apparent in the raw, high-dimensional space.
    * **Data Visualization:** The reduced dimensions from PCA can be used to create visualizations, which can help communicate complex data insights more effectively.

The clustering results are well-maintained, and the visualization highlights the distinct segments identified by the algorithm.
  * **Impact on Clustering:** PCA simplifies the dataset, making it easier to interpret and visualize the clustering results. This can be particularly useful in high-dimensional datasets where traditional clustering methods might struggle with noise or redundant features.
