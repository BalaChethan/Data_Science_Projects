## Paul's original data pre-processing code:

````python
#Import necessary libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer

#Sample data
data = pd.DataFrame({
    'Age': [25, 30, None, 35, 28],
    'Income': [50000, 60000, 75000, None, 55000],
    'Gender': ['M', 'F', 'M', 'F', 'M'],
    'Loan_Status': ['Approved', 'Rejected', 'Approved', 'Approved', 'Rejected']
})

print("Original Data: ")
print(data)

#Handling missing values with mean imputation
imputer = SimpleImputer(strategy='mean')
data[['Age', 'Income']] = imputer.fit_transform(data[['Age', 'Income']])

print("\nMissing Data replaced with mean: ")
print(data)

#Encoding categorical variables (Gender and Loan_Status)
le = LabelEncoder()
data['Gender'] = le.fit_transform(data['Gender'])
data['Loan_Status'] = le.fit_transform(data['Loan_Status'])

print("\n1-hot encoding categorical data: ")
print(data)

#Scaling numerical features (Age and Income) using StandardScaler
scaler = StandardScaler()
data[['Age', 'Income']] = scaler.fit_transform(data[['Age', 'Income']])

#Display the preprocessed and cleansed data
print("\nScaling numnerical features using StandardScaler: ")
print(data)
````

````python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
````

````python
# Load the dataset with specific columns
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/accepted_2007_to_2018Q4.csv', low_memory=False,
                   usecols=['id', 'loan_amnt', 'funded_amnt', 'int_rate', 'term', 'installment', 'grade', 'annual_inc', 'loan_status'])
````

````python
# Display the first 5 rows of the dataset for initial inspection
print("First 5 rows of the dataset:")
print(data.head())
````

````python
# Display statistics of numerical columns
print("\nStatistics for numerical columns:")
print(data[['loan_amnt', 'funded_amnt', 'int_rate', 'installment', 'annual_inc']].describe())
````

````python
# Check for missing values
print("\nMissing Values Count:")
print(data.isnull().sum())
````

````python
# Handle missing values for numerical columns using SimpleImputer (mean strategy)
imputer = SimpleImputer(strategy='mean')
data[['loan_amnt', 'funded_amnt', 'int_rate', 'installment', 'annual_inc']] = imputer.fit_transform(data[['loan_amnt', 'funded_amnt', 'int_rate', 'installment', 'annual_inc']])
````

````python
# Display statistics for categorical columns
print("\nStatistics for categorical columns:")
print(data.describe(include=['object']))
````

````python
# Plot histograms for numerical columns to visualize their distributions
numeric_columns = ['loan_amnt', 'funded_amnt', 'int_rate', 'installment', 'annual_inc']
data[numeric_columns].hist(bins=30, figsize=(15, 10))
plt.suptitle('Histograms of Numerical Columns', fontsize=16)
plt.show()
````

````python
# Plot count plots for categorical columns
categorical_columns = ['term', 'loan_status', 'grade']
plt.figure(figsize=(20, 10))
for i, col in enumerate(categorical_columns):
    plt.subplot(2, 2, i+1)
    sns.countplot(data[col])
    plt.title(f'Count Plot of {col}', fontsize=14)
plt.tight_layout()
plt.show()
````

````python
# Encode categorical variables using LabelEncoder and display label mappings
le = LabelEncoder()
for col in categorical_columns:
    data[col] = le.fit_transform(data[col])
    # Print label mappings
    print(f"\nLabel mappings for {col}:")
    for label, value in zip(le.classes_, le.transform(le.classes_)):
        print(f"  {label}: {value}")
````

````python
# Display encoded categorical variables
print("\nCategorical Variables After Encoding:")
print(data[categorical_columns].head())
````

````python
# Implement scaling on numerical features using StandardScaler
scaler = StandardScaler()
data[['loan_amnt', 'funded_amnt', 'int_rate', 'installment', 'annual_inc']] = scaler.fit_transform(data[['loan_amnt', 'funded_amnt', 'int_rate', 'installment', 'annual_inc']])
````

````python
# Display the final preprocessed dataset
print("\nFinal Preprocessed Data:")
print(data.head())
````
